```python
import google.generativeai as genai
import os
from google.colab import userdata

os.environ["GOOGLE_API_KEY"] = userdata.get('google_api_key')
# Configure the generative AI library with your API key
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.prompts import PromptTemplate, ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# Define llm
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0,
    verbose=True
)
```

```python
# Define tools and agent
!pip install wikipedia
from langchain.agents import load_tools, initialize_agent
tools = load_tools(["llm-math", "wikipedia"], llm=llm)
agent = initialize_agent(
    tools,
    llm,
    agent="zero-shot-react-description",
    handle_parsing_errors=True,
    verbose=True
)
```

```python
agent("Among the three famous cloud services providers: AWS from Amazon, Azure from Microsoft and Google Cloud, which one is the best and what's the reason?")
# Duanshuidashi
```

```python
# Define another agent using python REPL tool
from langchain_experimental.agents.agent_toolkits import create_python_agent
from langchain_experimental.tools.python.tool import PythonREPLTool
agent = create_python_agent(
    llm = llm,
    tool = PythonREPLTool(), # A python notebook like jupyter
    verbose = True
)
```

```python
agent.run("What is the 314th permutation of 'skelgmxlsk' in lexicographical order ?")
```

```python
# Create customized tools by define functions
from langchain.agents import tool
from datetime import date
@tool
def time(text:str)->str:
  """return today's date. \
  Use this for any questions related to\
  knowing today's date. Input should always\
  be an empty string. Any calculate should \
  occur outside of this function. """
  return str(date.today())

tools += [time]
agent = initialize_agent(
    tools,
    llm,
    agent="zero-shot-react-description",
    handle_parsing_errors=True,
    verbose=True
)
```

```python
agent("What's the date of today")
```
